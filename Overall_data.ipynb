# TEXT PRE_PROCESSING 

```python
import string
from nltk.corpus import stopwords
```

```python
def text_process(mess):
    """
    Takes in a string of text, then performs the following:
    1. Remove all punctuation
    2. Remove all stopwords
    3. Returns a list of the cleaned text
    """
    # Check characters to see if they are in punctuation
    nopunc = [char for char in mess if char not in string.punctuation]

    # Join the characters again to form the string.
    nopunc = ''.join(nopunc)
    
    # Now just remove any stopwords
    return [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]

mess = df['Public_Opinion_Vaccine']
```

```python
text_process(mess)
```

# Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
Bag_of_words = CountVectorizer(analyzer=text_process).fit(df['Public_Opinion_Vaccine'])
```

```python
# Print total number of vocab words
print(len(Bag_of_words.vocabulary_))
```

```python
BOW = Bag_of_words.transform(df['Public_Opinion_Vaccine'])
print('Shape of Sparse Matrix: ', BOW.shape)
print('Amount of Non-Zero occurences: ', BOW.nnz)
```

```python
sparsity = (100.0 * BOW.nnz / (BOW.shape[0] * BOW.shape[1]))
print('sparsity: {}'.format(round(sparsity)))
```

# Creating a Data Pipeline

```python
X = df['Public_Opinion_Vaccine']
y = df['Emotion']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size =0.3)
```

```python
from sklearn.pipeline import Pipeline

pipe = Pipeline([
    ('bow', CountVectorizer(analyzer=text_process)),  # strings to token integer counts
    ('tfidf', TfidfTransformer()),  # integer counts to weighted TF-IDF scores
    ('classifier', DecisionTreeClassifier()),  # train on TF-IDF vectors w/ Naive Bayes classifier
])
```

```python
pipe.fit (X_train,y_train)
```

```python
predictions = pipe.predict(X_test)
```

```python
predictions
```

```python
print (classification_report(predictions, y_test))
```
